## edit01

```python

# === General Setup ===
select solvent, solvent
select Na, resn NA+CL
remove solvent
remove Na
remove (hydro); 
bg_color white
#hide all
hide everything
# === Select Molecular Parts ===
select Pep_neg, resn ALA+ASP+GLN+GLU+GLY+PHE+TRP+TYR+VAL
select DNA, resn DA+DT+DC+DG+TO3+TO5


select bases, resn DA+DC+DT+DG+DG3+DG5+DT3+DT5+DA3+DA5+DC3+DC5+TO3+TO5 and name c2+c4+c5+c6+c7+c8+n1+n2+n3+n4+n6+n7+n9+o2+o4+o6+c5m; 
select backbone_, resn DA+DC+DT+DG+DG3+DG5+DT3+DT5+DA3+DA5+DC3+DC5+TO3+TO5 and name o1p+o2p+op1+op2+p+c1'+c2'+c3'+c4'+c5'+o3'+o4'+o5'; 

select chainA, resi 1-413 and resn DA+DC+DT+DG+DG3+DG5+DT3+DT5+DA3+DA5+DC3+DC5+TO3+TO5


#nonmodified chain
select chainB, resi 414-560; 
select chainA, chainA and backbone_; 
select chainB, chainB and backbone_; 

#####---------------------DNA----------------------#####
set_color myblue, [ 0.106, 0.627, 1 ]; 
color myblue, chainA; 
color myblue, chainB; 
cartoon oval; 
set cartoon_ring_mode, 1; 
set cartoon_ring_transparency, 0.5;
set cartoon_ring_width, 0.7; 
color gray90, bases;
bg_color white;
set fog, off; 
set cartoon_oval_width, 0.2; 
set cartoon_oval_length, 1.3; 
set cartoon_ring_finder, 3
show cartoon, DNA
# color myblue, DNA
#####---------------------linkers----------------------#####
select linker_TO5, resn TO5 and  name O9L+C27L+C26L+C10L+C25L+C24L+C23L+S22L+C15L+C11L+N12L+C13L+C14L+O17L+O16L+C18L+C19L+C20L+C21L+C49L+C+O56L+O1P+O2P+P
select linker_TO3, resn TO3 and  name O2L+C16L+C15L+C14L+S13L+C6L+C5L+C4L+O8L+N3L+C1L+O7L+C9L+C10L+C11L+C12L+C17L+C+O39L
# select alkyl_chains_connect, alkyl_chains_connect or name oa;
show sticks, linker_TO3;
# set stick_radius, 0.6;
set_color mygray, [0.35, 0.35, 0.35];
color mygray, linker_TO5;
color mygray, linker_TO3;
show licorice, linker_TO5;
hide cartoon, linker_TO5;
hide cartoon, linker_TO3;
show licorice, resn TO3;
show licorice, resn TO5;
# set sphere_scale, 0.6




#####---------------------Lipids----------------------#####
select lipids, resn DMPC
# select hydrophilic_DMPC, resn DMPC and name C21+O22+O21+C2+C31+O32+O31+C3+HX+HY+HS+O11+P+O14+O13+O12+C11+C12+H12B+H12A+H11A+H11B+N+C15+H15A+H15C+H15B+H14A+H14B+H14C+C14+C13+H13A+H13B+H13C+C1+HB+HA
# select  hydrophobic_DMPC, resn DMPC and name H14Z+C22+H2R+H2S+C23+C24+C25+C26+C27+C28+C29+C210+C211+C212+C213+C214+C314+C313+C312+C311+C310+C39+C38+C37+C36+C35+C34+C33+C32+H3R+H3S+H4S+H4R+H5R+H5S+H6R+H6S+H7R+H7S+H8R+H8S+H9R+H9S+H10R+H10S+H11R+H11S+H12R+H12S+H13R+H13S+H14R+H14S+H14T+H2X+H2Y+H3X+H3Y+H4X+H4Y+H5X+H5Y+H6X+H6Y+H7X+H7Y+H8X+H8Y+H9X+H9Y+H10X+H10Y+H11X+H11Y+H12X+H12Y+H13X+H13Y+H14X+H14Y+H14T


#color lipids
#-> first option
# set_color sand, [0.85, 0.8, 0.7]
# color sand, lipids
#-> second option 
# Lipid
color gray90,  lipids
show surface, lipids
set transparency, 0.35, lipids
set surface_quality, 1
set specular, 0.2


#####----------------------Neg4F----------------------#####
# color Neg4F
set_color peptide_neg_col, [0x94, 0x21, 0x92]
color peptide_neg_col,  Pep_neg
hide everything, resi 738+742+734

#representation Neg4F
show cartoon, Pep_neg
show licorice, Pep_neg
```
